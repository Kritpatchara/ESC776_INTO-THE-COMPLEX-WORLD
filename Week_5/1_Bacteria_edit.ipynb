```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# 1. Virtual population initiation

n_initial_pop = 100000
initial_pop = np.random.normal(100, 15, n_initial_pop) # np.random.normal(mean, SD, n)
plt.hist(initial_pop, np.arange(0, 200, 1)) # plt.hist(data, bins)
```

```python
# 2. Natural selection

drug = 120
selected_pop = []
for i in initial_pop:
    if i >= drug:
        selected_pop.append(i)
plt.hist(initial_pop, np.arange(0, 200, 1));
plt.hist(selected_pop, np.arange(0, 200, 1));
```

```python
# 3. Reproduction
reproduced_pop = np.random.choice(selected_pop, n_initial_pop)
plt.hist(initial_pop, np.arange(0, 200, 1));
plt.hist(reproduced_pop, np.arange(0, 200, 1));
plt.hist(selected_pop, np.arange(0, 200, 1));
```

```python
# 3. Mutation
mutated_pop = []
for i in reproduced_pop:
    r = np.random.choice([-3, -2, -1, 0, 1, 2, 3])
    mutated_pop.append(i+r)
plt.hist(initial_pop, np.arange(0, 200, 1));
plt.hist(mutated_pop, np.arange(0, 200, 1));
plt.hist(selected_pop, np.arange(0, 200, 1));
'''
# Next generation
initial_pop = mutated_pop
'''
```
