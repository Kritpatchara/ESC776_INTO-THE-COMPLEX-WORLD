a + 2b + 3c = 20
Find a, b, c
Where a >= 0, b >= 0, c >= 0

```python
import numpy as np
```

```python
# Concept

# fitness score = |ai + 2bi + 3ci - 20|
# We want fitness score -> 0
# We select only cells with fitness score <= average
```

```python
# Initial population creation
initial_pop = []
n = 5
for i in range(n):
    a = np.random.randint(0, 20)
    b = np.random.randint(0, 20)
    c = np.random.randint(0, 20)
    fitness_score = abs(a+2*b+3*c-20)
    initial_pop.append([a, b, c, fitness_score])
```

```python
# Selection
averagefit = np.average(np.array(initial_pop)[:, 3])

selected_pop = []
for i in initial_pop:
    if i[3] <= averagefit:
        selected_pop.append(i)
    if i[3] == 0:
        print(i)
```

```python
# Reproduction
index = np.random.randint(len(selected_pop), size = n)
reproduced_pop = np.array(selected_pop)[index,:]
```

```python
# Mutation
mutated_pop = []
for i in reproduced_pop:
    random_a = np.random.choice([-1, 0, 1])
    random_b = np.random.choice([-1, 0, 1])
    random_c = np.random.choice([-1, 0, 1])
    a = abs(i[0]+random_a)
    b = abs(i[1]+random_a)
    c = abs(i[2]+random_a)
    fitness_score = abs(a+2*b+3*c-20)
    mutated_pop.append([a, b, c, fitness_score])

initial_pop = mutated_pop # for next generation
```

```python
# Next generation
for generation in range(100):
    # Selection
    averagefit = np.average(np.array(initial_pop)[:, 3])

    selected_pop = []
    for i in initial_pop:
        if i[3] <= averagefit:-
            selected_pop.append(i)
        if i[3] == 0:
            print(i)

    # Reproduction
    index = np.random.randint(len(selected_pop), size = n)
    reproduced_pop = np.array(selected_pop)[index,:]

    # Mutation
    mutated_pop = []
    for i in reproduced_pop:
        random_a = np.random.choice([-1, 0, 1])
        random_b = np.random.choice([-1, 0, 1])
        random_c = np.random.choice([-1, 0, 1])
        a = abs(i[0]+random_a)
        b = abs(i[1]+random_a)
        c = abs(i[2]+random_a)
        fitness_score = abs(a+2*b+3*c-20)
        mutated_pop.append([a, b, c, fitness_score])

    initial_pop = mutated_pop # for next generation
```
