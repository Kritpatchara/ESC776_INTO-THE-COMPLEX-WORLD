```python
import numpy as np
import matplotlib.pyplot as plt
```

You: 100 Baht, 
House: 2000 Baht

Head: get 10 Baht, 
Tail: lose 10 Baht

Play until either you or the house is bankrupt.

1. Plot sample path in a game.
2. In 500 games, plot histogram of #trials per game. How many game do you win? How many game does house win?

```python
you = 100
your_money = [100]
house = 2000
house_money = [2000]
trial = 0

while you > 0 and house > 0:
    i = np.random.randint(1, 3) # 1 = head, 2 = tail
    if i == 1:
        you += 10
        house -= 10
    else:
        you -= 10
        house += 10
    your_money.append(you)
    house_money.append(house)
    trial += 1
print(trial)
plt.plot(your_money)
plt.plot(house_money)
plt.show()
```

```python
trials = []
your_money = [100]
house_money = [2000]
house_win = 0
you_win = 0
for i in range(500):
    you = 100
    house = 2000
    trial = 0
    while you > 0 and house > 0:
        i = np.random.randint(1, 3) # 1 = head, 2 = tail
        if i == 1:
            you += 10
            house -= 10
        else:
            you -= 10
            house += 10
        your_money.append(you)
        house_money.append(house)
        trial += 1
    if you > 0:
        you_win += 1
    elif house > 0:
        house_win += 1
    trials.append(trial)


plt.hist(trials)
plt.show()
print("House win: ", house_win)
print("You win: ", you_win)
```
