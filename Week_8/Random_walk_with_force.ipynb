```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Condition

n_step = 1000
n_walker = 2000
dt = 10
m = 32.0*1.67e-27 # Oxygen mass in kg
k_B = 1.38e-23 # Boltzmann constant
T = 300 # temperature in Kelvin
speed = 1e-3*np.sqrt(3*k_B*T/m) # V_rms
g = 0.00981
max_stepsize = speed*dt
```

```python
# Atmosphere

H0 = 10 # troposphere thickness = 10 km
pos = np.zeros((n_step+1, n_walker)) # position array of many walkers
pos[0] = H0*np.random.rand(n_walker) # randomize position of O2 in troposphere

for i in range(n_step) :
    theta = 2*np.pi*np.random.rand(n_walker) # randomize polar angles
    # positions in y-axis - positive only (above ground) + gravity
    pos[i+1] = np.fabs(pos[i] + max_stepsize*np.sin(theta) - 0.5*g*dt**2)

# show results - halfway & final - checking equilibration
print('start')
plt.ylim(top=100)
plt.plot(pos[0], '.')
plt.show()

plt.ylim(top=100)
print('halfway')
plt.plot(pos[int(n_step/2)],'.')
plt.show()

plt.ylim(top=100)
print('final')
plt.plot(pos[n_step],'.')
plt.show()
```
