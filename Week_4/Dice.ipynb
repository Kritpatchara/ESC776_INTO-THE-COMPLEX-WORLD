```python
import numpy as np
import matplotlib.pyplot as plt
```

Paris Salon Problem:
Would you bet 10 Baht to make 10 Baht that a 6 will not be obtained in 4 rolls of a fair die?

Answer: I will play this game. I think five from six number have more chance to win than only one number.

```python
# 1. Using Monte Carlo methods to simulate 1,000 games (initial money is 0), plot a sample path to show the remaining balance in each trial.
money = 0
money_timeline = [0]
expected_value_timeline = []
for i in range(1000):
    dice = []
    for i in range(4):
        dice.append(np.random.randint(1,7))
    if 6 in dice:
        money -= 10
    else:
        money += 10
    money_timeline.append(money)
    expected_value = money/(len(money_timeline)-1)
    expected_value_timeline.append(expected_value)

print(money) # Find an expected value of this game (per a trial): final remaining money / 1,000
print(expected_value)
print(max(expected_value_timeline))
plt.plot(money_timeline)
plt.show()
plt.hist(expected_value_timeline)
plt.show()
```
