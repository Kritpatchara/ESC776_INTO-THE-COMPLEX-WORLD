```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1 walkers in 100 steps
n_step = 100
position = np.zeros(n_step+1) # position array
position[0] = 0. # initial position

for i in range(n_step):
    direction = np.random.rand()
    if direction > 0.5: # decision
        position[i+1] = position[i]+1
    else:
        position[i+1] = position[i]-1
plt.plot(position)
```

```python
# 20 walkers in 10 steps
n_step = 500
n_walker = 200
position = np.zeros((n_step+1, n_walker)) # rows, x-axis for step and column, y-axis is walker

for i in range(n_step):
    # Random direction
    r = np.random.rand(n_walker) # Random a number
    r1 = np.floor(2*r) # Greater than 1 or less than 1
    direction = 2*r1-1

    position[i+1] = position[i] + direction

plt.imshow(position)
plt.colorbar()
plt.xlabel("Walker")
plt.ylabel("Position")
plt.show()

for j in range(n_walker):
    plt.plot(position[:,j])
plt.show()
```

```python
# Distribution - Histogram - Variance
# fig, ax = plt.subplots()
# ax.set_ylim(120)
# ax.hist(position[20], range = (-60, 60), bins = 20) # Distribution of timestep
# plt.show()

# Animation
for i in [5, 10, 20, 50, 100, 200, 499]:
    print("time = %d, variance = %.2f" % (i, np.var(position[i])))
    
    fig, ax = plt.subplots()
    ax.set_ylim(0, 120)  # Adjusted ylim to start from 0
    ax.hist(position[i], range=(-60, 60), bins=20)  # Changed orientation to "vertical"
    
    plt.title("Histogram for time = %d" % i)
    plt.xlabel("Position")
    plt.ylabel("Frequency")
    plt.show()
```
